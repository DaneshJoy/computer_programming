### Class Attributes and `__init__` Method

```python
class Patient:
    def __init__(self, name, age, heart_rate, blood_pressure): 
        self.name = name
        self.age = age
        self.heart_rate = heart_rate
        self.blood_pressure = blood_pressure
        self.status = 'پذیرش شده'
```

### Class Methods

```python
class Patient:
    def __init__(self, name, age, heart_rate, blood_pressure): 
        self.name = name
        self.age = age
        self.heart_rate = heart_rate
        self.blood_pressure = blood_pressure
        self.status = 'پذیرش شده'

    def show_info(self):
        print(f'Patient Name: {self.name}')
        print(f'HR: {self.heart_rate}')
        print(f'BP: {self.blood_pressure}')
        print(f'Status: {self.status}')

    def check_heart_rate(self):
        if self.heart_rate < 60:
            return 'برادیکاردی'
        elif self.heart_rate < 100:
            return 'نرمال'
        else:
            return 'تاکیکاردی'

    def full_report(self):
        print('=' * 35)
        self.show_info()
        print('Heart rate status:', self.check_heart_rate())
        print('=' * 35)
        
```

```python
patient1 = Patient('Ali Rezaee', 45, 78, 120)
patient2 = Patient('Maryam Ahmadi', 32, 92, 110)
```

```python
patient1.show_info()
```

```python
patient2.full_report()
```

## Inheritance

```python
class ICUPatient(Patient):
    pass
```

```python
patient3 = ICUPatient('Sara Karimi', 28, 72, 115)
patient4 = ICUPatient('Reza Motaharri', 67, 85, 88)
```

```python
class ICUPatient(Patient):
    def __init__(self, name, age, heart_rate, blood_pressure, oxygen_level):
        super().__init__(name, age, heart_rate, blood_pressure)
        self.oxygen_level = oxygen_level

    def check_oxygen(self):
        if self.oxygen_level < 90:
            return 'هیپوکسی'
        elif self.oxygen_level < 95:
            return 'مرزی'
        else:
            return 'نرمال'

    def full_report(self):
        super().full_report()
        print('Oxygen Level:', self.oxygen_level)

    def icu_report(self):
        self.show_info()
        print('Oxygen Level:', self.oxygen_level)
```

```python
patient3 = ICUPatient('Sara Karimi', 28, 72, 115, 90)
patient4 = ICUPatient('Reza Motaharri', 67, 85, 88, 83)
```

```python
patient3.show_info()
```

```python
patient4.full_report()
```

```python
patient3.icu_report()
```

## Check Class of an Object

```python
print(isinstance(patient3, Patient))
print(isinstance(patient3, ICUPatient))
```
