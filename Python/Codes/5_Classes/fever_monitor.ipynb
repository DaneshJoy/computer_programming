# Fever Monitor

```python
import numpy as np
import matplotlib.pyplot as plt
```

## PatientMonitor Class

```python
class PatientMonitor:
    def __init__(self, name, hours, temperatures):
        self.name = name
        self.hours = hours
        self.temperatures = temperatures

    def statistics(self):
        temp_mean = np.mean(self.temperatures).round(2)
        temp_min = np.min(self.temperatures).round(2)
        temp_max = np.max(self.temperatures).round(2)
        return temp_mean, temp_min, temp_max

    def report(self):
        print(f'Patient Name: {self.name}')
        mean_, min_, max_ = self.statistics()
        print(f'Statistics:\n\tMean: {mean_}\n\tMin: {min_}\n\tMax: {max_}')

    def plot_temperatures(self, figsize=(5,3)):
        plt.figure(figsize=figsize)
        plt.plot(self.hours, self.temperatures, 'b--', marker='o')
        plt.title(f'{self.name} temperatures')
        plt.xlabel('Hour')
        plt.ylabel('Temperature')
        plt.grid(True)
        plt.show()
```

```python
hours = np.arange(24)
temperatures1 = np.random.normal(loc=37, scale=1, size=24).round(2)

print(hours)
print(temperatures1)
```

```python
patient1 = PatientMonitor('Saeed',
                          hours=hours,
                          temperatures=temperatures1)
```

```python
patient1.report()
```

```python
patient1.plot_temperatures(figsize=(7,2))
```
