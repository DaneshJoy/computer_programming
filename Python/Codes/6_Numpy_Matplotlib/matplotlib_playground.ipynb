# Matplotlib Playground

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create two signals

```python
x = np.linspace(0, 2*np.pi, 500)
y1 = np.sin(x)
y2 = 3 * np.sin(2*x)
```

```python
print(x.shape)
print(y1.shape)
print(y2.shape)
```

## Simple Plot

```python
plt.figure(figsize=(5,3))
plt.plot(x, y1)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sin (x)')
plt.grid(True)
plt.show()
```

## Multiple plots on the same figure

```python
# supported color letters: r, g, b, c, m, y, k
# supported styles: -, --, -., :, etc.

plt.figure(figsize=(5, 3))

plt.plot(x, y1, 'm--', label='Sin(x)')
plt.plot(x, y2, 'g:', label='3*Sin(2x)')

plt.legend()

plt.savefig('fig.png')

plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(6, 2), sharey=True)

axes[0].plot(x, y1)
axes[0].set_title('Sin(x)')
axes[0].set_xlabel('time')
axes[0].set_ylabel('amplitude')
axes[0].set_ylim([-3, 3])

axes[1].plot(x, y2)
axes[1].set_title('3*Sin(2x)')
axes[1].set_xlabel('time')
axes[1].set_ylim([-3, 3])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(5, 4), sharex=True)

axes[0].plot(x, y1, 'c--')
axes[0].set_title('Sin(x)')
axes[0].set_ylabel('amplitude')

axes[1].plot(x, y2, 'm-.')
axes[1].set_title('3*Sin(2x)')
axes[1].set_ylabel('amplitude')

plt.tight_layout()
plt.show()
```
