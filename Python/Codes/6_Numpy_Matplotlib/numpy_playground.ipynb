## توابع ایجاد آرایه

```python
print(np.zeros(5))
print('-'*30)
print(np.zeros([3, 4]))
print('-'*30)
print(np.ones(5))
print('-'*30)
print(np.ones([3, 4]))
```

```python
print(np.random.random(5))
print('-'*30)
print(np.random.random([2, 3]))
print('-'*30)
print(np.random.randint(0, 6, [4, 4]))  # (low, high, size)
```

```python
print(np.arange(1, 10))  # (start, stop)
print('-'*30)           
print(np.arange(1, 20, 2))  # (start, stop, step)
print('-'*30)
print(np.arange(0, 1, 0.1))
print('-'*30)
print(np.linspace(0, 1, 9))  # (start, stop, num)
```

```python
print(np.full([3, 4], 5))  # (shape, value)
```

```python
mat_ones = np.ones([3, 4])
mat_ones
```

```python
mat1 = mat_ones * np.array([1, 2, 3, 4])
mat1
```

```python
np.array([1, 2, 3]).reshape(-1, 1)
```

```python
mat2 = mat_ones * np.array([1, 2, 3]).reshape(-1, 1)
mat2
```

## محورها (Axis)

```python
print(np.sum(mat1))
print(np.sum(mat1, axis=0))
print(np.sum(mat1, axis=1))
```

```python
print(np.mean(mat1))
print(np.mean(mat1, axis=0))
print(np.mean(mat1, axis=1))
```

## عملیات برداری و ماتریسی

```python
vec1 = np.array([1, 2, 3])
vec2 = np.array([2, 2, 2])
```

```python
print(vec1 * vec2)
print(np.multiply(vec1, vec2))
```

```python
# Dot product
print(vec1 @ vec2)
print(np.dot(vec1, vec2))
```

```python
# Cross product
print(np.cross(vec1, vec2))
```

```python
mat_1 = np.full([3, 4], 2)
mat_2 = np.full([3, 4], 5)
```

```python
print(mat_1 * mat_2)
print('-'*30)
print(np.multiply(mat_1, mat_2))
```

```python
print(mat1 @ np.transpose(mat_2))
print('-'*30)
print(np.matmul(mat1, np.transpose(mat_2)))
```

## Conditional Indexing / Filtering

```python
arr = np.arange(1, 10)
arr
```

```python
# Greater than 5
arr > 5
```

```python
# Greater than 5
arr[arr > 5]
```

```python
# Greater than 2 AND Less than 6  (2<x<6)
arr[(arr>2) & (arr<6)]
```

```python
# Less than 2 OR Greater than 6  (x<2 or x>6)
arr[(arr<2) | (arr>6)]
```

```python
# number of items greater than 5
print(np.sum(arr > 5))
```

```python
# items that are greater than 5 (where)
print(np.where(arr > 5))
```

## Roots of a polynomial function

```python
# x^2 - 5X + 6 = (x-2)(x-3) => roots: 2, 3
np.roots([1, -5, 6])
```

```python
# 3x^3 - 5x + 5
np.roots([3, 0, 5, 5]).round(2)
```
